```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
```

```python
#create dataset
```

```python
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=123)
```

```python
X
```

```python
pd.DataFrame(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
regression=LogisticRegression()
regression.fit(X_train,y_train)
y_pred=regression.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import precision_score,accuracy_score,confusion_matrix,fbeta_score,recall_score,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
fbeta_score(y_test,y_pred,beta=1)
```

```python
recall_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# hyperparameter tuning
```

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1,0.1,0.2,0.01]
solver=['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
```

```python
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
#gridsearchcv
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv)
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Randomized SearchCV
```

```python
from sklearn.model_selection import RandomizedSearchCV
random=RandomizedSearchCV(estimator=model,param_distributions=params,cv=5,scoring='accuracy')
```

```python
random.fit(X_train,y_train)
```

```python
y_pred=random.predict(X_test)
```

```python
random.best_score_
```

```python
random.best_params_
```

```python
accuracy_score(y_test,y_pred)
```

```python
classification_report(y_test,y_pred)
```

```python
# Logistic regression for muticlass classification problem
```

```python
X,y=make_classification(n_samples=1000,n_features=10,n_classes=3,random_state=123,n_informative=3)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=33)
```

```python
from sklearn.linear_model import LogisticRegression
logistic= LogisticRegression(multi_class='ovr')
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
